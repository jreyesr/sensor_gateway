## Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

## Load data

Load measurement data. Columns are Location (an ID code), Date, VW_30cm and T_30cm

```python
df = pd.read_csv('./data.tsv', sep='\t', header=0, usecols=[0, 1, 2, 7]).dropna()
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
df["Location"] = df["Location"].astype(str)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
locs = pd.read_csv("./sensor_locs.tsv", sep="\t", header=1)
locs["Location"] = locs["Location"].astype(str)
```

```python
print(locs.info())
```

```python
print(locs.head())
```

Join location data with measurements

```python
meas = df.set_index("Location").join(locs.set_index("Location"))
```

```python
print(meas.head())
```

```python
latest = meas.sort_values("Date").groupby(["Location"]).tail(1)
print(latest.head())
```

## Interpolation with scipy.interpolate.Rbf

Get extents (min/max Easting & Northing)

```python
extent = x_extent = x_min, x_max, y_min, y_max = [latest.Easting.min()-100, latest.Easting.max()+100,
                                                  latest.Northing.min()-100, latest.Northing.max()+100]
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(latest.Easting, latest.Northing, c=latest.T_30cm)
ax.set_aspect(1)
ax.set_xlim(*extent[:2])
ax.set_ylim(*extent[2:])
ax.set_xlabel('Easting [m]')
ax.set_ylabel('Northing [m]')
ax.set_title('Temp @ 30cm depth (ºC)')
ax.grid(c='k', alpha=0.2)

plt.show()
```

Create resampling grid (uniformly spaced at GRID_SPACING meters)

```python
NUM_SAMPLES_X, NUM_SAMPLES_Y = 512, 400
GRID_SPACING_X = (x_max-x_min)/NUM_SAMPLES_X
GRID_SPACING_Y = (y_max-y_min)/NUM_SAMPLES_Y
print(GRID_SPACING_X, GRID_SPACING_Y)
grid_x, grid_y = np.mgrid[x_min:x_max:GRID_SPACING_X, y_min:y_max:GRID_SPACING_Y]

plt.figure(figsize=(10,6))
plt.scatter(grid_x, grid_y, s=10)
```

```python
from scipy.interpolate import Rbf
# Make an n-dimensional interpolator
rbfi = Rbf(latest.Easting, latest.Northing, latest.T_30cm)
# Predict on the regular grid
di = rbfi(grid_x, grid_y)
```

```python
mi = np.min(np.hstack([di.ravel(), latest.T_30cm.values]))
ma = np.max(np.hstack([di.ravel(), latest.T_30cm.values]))
plt.figure(figsize=(15,15))
c1 = plt.imshow(di.T, origin="lower", extent=extent, vmin=mi, vmax=ma)
c2 = plt.scatter(latest.Easting, latest.Northing, s=60, c=latest.T_30cm, edgecolor='#ffffff66', vmin=mi, vmax=ma)
plt.colorbar(c1, shrink=0.55)
plt.show()
```

```python
plt.imsave("colormap.png", di.T, origin="lower", vmin=mi, vmax=ma)
print([x_min, y_min, x_max, y_max])
```

```python
por_hat = rbfi(latest.Easting, latest.Northing)
sns.distplot(por_hat - latest.T_30cm)
plt.show()
```

## Interpolation with `scipy.griddata()`

```python
points = latest[['Easting', 'Northing']].values
```

```python
from scipy.interpolate import griddata

grid_z0 = griddata(points, latest.T_30cm.values, (grid_x, grid_y), method='nearest')
grid_z1 = griddata(points, latest.T_30cm.values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, latest.T_30cm.values, (grid_x, grid_y), method='cubic')
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(15, 5))

ax = axs[0]
ax.imshow(grid_z0, origin='lower', extent=extent)
ax.scatter(latest.Easting, latest.Northing, s=2, c='w')
ax.set_title('Nearest')

ax = axs[1]
ax.imshow(grid_z1, origin='lower', extent=extent)
ax.scatter(latest.Easting, latest.Northing, s=2, c='w')
ax.set_title('Linear')

ax = axs[2]
ax.imshow(grid_z2, origin='lower', extent=extent)
ax.scatter(latest.Easting, latest.Northing, s=2, c='w')
ax.set_title('Cubic')

plt.show()
```

## Using `sklearn.gaussian_process`

```python
from sklearn.gaussian_process.kernels import RBF

kernel = RBF(length_scale=100)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

gp = GaussianProcessRegressor(normalize_y=True,
                              alpha=0.1,  # Larger values imply more noise in the input data.
                              kernel=kernel,)

gp.fit(latest[['Easting', 'Northing']].values, latest.T_30cm.values)
```

```python
X_grid = np.stack([grid_x.ravel(), grid_y.ravel()]).T
```

```python
y_grid = gp.predict(X_grid).reshape(grid_x.shape)
```

```python
# Compute min and max of all the data:
mi = np.min(np.hstack([y_grid.ravel(), latest.T_30cm.values]))
ma = np.max(np.hstack([y_grid.ravel(), latest.T_30cm.values]))

# Plot it all.
plt.figure(figsize=(15,15))
im = plt.imshow(y_grid.T, origin='lower', extent=extent, vmin=mi, vmax=ma)
pts = plt.scatter(latest.Easting, latest.Northing, c=latest.T_30cm, s=80, edgecolor='#ffffff66', vmin=mi, vmax=ma)
plt.colorbar(im, shrink=0.55)
plt.show()
```

```python
por_hat = gp.predict(latest[['Easting', 'Northing']].values)

sns.distplot(por_hat - latest.T_30cm)
```

## Using `verde.Spline`

```python
import verde as vd

extent = x_min, x_max, y_min, y_max = vd.pad_region(vd.get_region((latest.Easting, latest.Northing)), pad=100)
```

```python
spline = vd.Spline(mindist=2000, damping=1e-2)
spline.fit((latest.Easting, latest.Northing), latest.T_30cm)
```

```python
grid = spline.grid(region=extent, spacing=min(GRID_SPACING_X, GRID_SPACING_Y))
grid
```

```python
mi = np.min(np.hstack([grid.scalars.values.ravel(), latest.T_30cm.values]))
ma = np.max(np.hstack([grid.scalars.values.ravel(), latest.T_30cm.values]))

# Plot it all.
plt.figure(figsize=(15,15))
im = plt.imshow(grid.scalars, origin='lower', extent=extent, vmin=mi, vmax=ma)
pts = plt.scatter(latest.Easting, latest.Northing, c=latest.T_30cm, s=80, edgecolor='#ffffff66', vmin=mi, vmax=ma)
plt.colorbar(im, shrink=0.55)
plt.show()
```
